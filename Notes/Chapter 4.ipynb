```python
from tsfeatures import *
import polars as pl 
import numpy as np
import os
from dotenv import load_dotenv
from matplotlib import pyplot as plt
import seaborn as sns
from lets_plot import *
LetsPlot.setup_html()

pl.Config.set_tbl_rows(100)

# Load environment variables from .env file
load_dotenv()
# Retrieve the dataset path from the environment variable
DATASET = os.getenv('DATASET')
df = pl.read_parquet(DATASET)
```

```python
# Aggregate
def agg_dengue(df,group=''):
    if group == '':
        cols = ['date','epiweek','week','year','season','season_week']
    else:
        cols = ['date','epiweek','week','year','season','season_week',group]
    gdf = df.group_by(cols).agg(
        pl.col('casos').sum().alias('casos'),
        pl.col('population').mean().alias('population'),
        pl.col('outlier').first().alias('outlier'),
    ).sort('year')
    
    gdf = gdf.with_columns(
        ((pl.col('casos')/pl.col('population')) *100000).alias('dir'),
        np.log1p(pl.col('casos')).alias('log_casos'),
        np.log1p(((pl.col('casos')/pl.col('population')) *100000)).alias('log_dir'),
    )
    return gdf

gdf = agg_dengue(df, group='uf')
```

```python
stl_feat = tsfeatures(ts, freq=53, features=[stl_features])

stl_feat
```

```python
(
    ggplot(data=stl_feat)
    + geom_point(aes(x='trend',y='seasonal_strength',color='unique_id'))
)
```

Note that 'SC' and 'ES' have the largest trend, but, 'ES' has missing data. Also, 'PB' has lowest trend.
On the other hand, 'RO' has largest seasonality, while 'MG' has lowest seasonality.
Let us visualize them.

```python
p1 = (
    ggplot(data=gdf.filter(pl.col('uf').is_in(['PB','SC'])))
    + geom_line(aes(x='date',y='log_casos',color='uf'))
    + facet_grid(x="uf")
)
p2 = (
    ggplot(data=gdf.filter(pl.col('uf').is_in(['CE','GO'])))
    + geom_line(aes(x='date',y='log_casos',color='uf'))
    + facet_grid(x="uf")
)


gggrid([p1, p2], ncol=1)
```

We can clearly see the difference in terms of trends. Indeed, 'SC' has a clear growing trend.
On the other had, the seasonality is not clearly observed from the graphs.
Why is 'RO' the most seasonal? The reason is actually that 'RO' has almost zero cases, hence,
the number of cases is mostly driven by the "seasonal" effect. For example, if we have 0 cases all year long,
except one month where we have 100 cases, then our time serie is mostly driven by seasonal effects.
Therefore, our measure of seasonality is not invariant under the absolute number of cases, which
makes it less useful in terms of capturing the relative effect.
