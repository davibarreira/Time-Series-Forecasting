### 1 - A single forecast

```python
import polars as pl 
import pandas as pd
import numpy as np
import os
from dotenv import load_dotenv
from matplotlib import pyplot as plt
import seaborn as sns
from lets_plot import *
LetsPlot.setup_html()

pl.Config.set_tbl_rows(100)

# Load environment variables from .env file
load_dotenv()
# Retrieve the dataset path from the environment variable
DATASET = os.getenv('DATASET')
df = pl.read_parquet(DATASET)

# Aggregate
def agg_dengue(df,group=''):
    if group == '':
        cols = ['date','epiweek','week','year','season','season_week']
    else:
        cols = ['date','epiweek','week','year','season','season_week',group]
    gdf = df.group_by(cols).agg(
        pl.col('casos').sum().alias('casos'),
        pl.col('population').mean().alias('population'),
        pl.col('outlier').first().alias('outlier'),
    ).sort('year')
    
    gdf = gdf.with_columns(
        ((pl.col('casos')/pl.col('population')) *100000).alias('dir'),
        np.log1p(pl.col('casos')).alias('log_casos'),
        np.log1p(((pl.col('casos')/pl.col('population')) *100000)).alias('log_dir'),
    )
    return gdf

def colapse_52_53(df):
    temp = df.filter(pl.col('week') == 53)[['epiweek','geocode','casos']].with_columns(
        (pl.col('epiweek')-1).alias('epiweek')
    )
    
    df = df.join(temp,on=['geocode','epiweek'],how='left').with_columns(
        pl.when(pl.col('casos_right').is_null())
        .then(pl.col('casos'))
        .otherwise(pl.col('casos_right') + pl.col('casos'))
        .alias('casos')
    ).drop('casos_right')

    df = df.filter(pl.col('week') < 53)
    return df

df = colapse_52_53(df)
gdf = agg_dengue(df)
gdf = gdf.filter(pl.col('year')>=2018,pl.col('season')!='2017-2018')
ts = gdf.to_pandas().copy()[['date','log_casos']].rename(columns = {'date':'ds','log_casos':'y'})
ts['unique_id'] = 'Brasil'

ts = ts.sort_values(by='ds').reset_index(drop=True)
```

```python
from statsforecast import StatsForecast
from statsforecast.models import (
    AutoARIMA,
    HistoricAverage,
    DynamicOptimizedTheta as DOT,
    SeasonalNaive,
    AutoETS,
    HoltWinters
)
```

```python
StatsForecast.plot(ts)
```

```python
models = [
    AutoARIMA(season_length = 52, seasonal=True, stepwise=True),
    SeasonalNaive(season_length=52),
    HistoricAverage(),
    DOT(season_length=52),
]

sf = StatsForecast(
    models=models,
    freq='W',
    n_jobs=-1
)
sf.fit(ts)

# Use this to see the fitted model paramaters
# sf.fitted_[0,0].model_
```

```python
forecast_df = sf.predict(h=52, level=[90])
forecast_df.tail()
```

```python
sf.plot(ts, forecast_df, level=[90])
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose 
a = seasonal_decompose(ts["y"], model = "additive", period=52)
a.plot();
```

### Cross Validation

```python
cv_df = sf.cross_validation(
    df = ts,
    h = 52,
    step_size = 52,
    n_windows = 4
  )
```
